```python
import json
import random
import requests, os, re, time
from time import sleep
# 一樣使用內建的 urllib.request 裡的 urlopen 這個功能來送出網址
from urllib.request import urlopen,urlretrieve,Request
# 在你剛剛安裝的 beautifulsoup4 函式庫裡使用 BeautifulSoup 這個解析器
from bs4 import BeautifulSoup
# 送出網址, 這次我送出的是網址列的網址, 回應則是一個網頁的形式

account_list = ["evenyiing", "gogo5885", "bprk6794","feelmi","miss_aprils"]

for account in account_list:
    page = 1
#     account = "top0988"
    print("商家帳號:", account)
    dn = "ruten/" + account + '/'
    
    while True:
        url = "https://class.ruten.com.tw/user/index00.php?s=" + account +"&p=" + str(page)
        print("第", page, "頁:", url)
        
        headers = {
            "user-agent": "Mozilla/5.0",
            "Sec-Fetch-Mode":"no-cors"
        }
        
        response = requests.get(url, headers=headers)
        # 把回應丟到解析器
        html = BeautifulSoup(response.text)
        
        
        shoes=html.find_all('div',class_='rt-store-goods-disp-mix')
        if shoes == []:
            print("共", page, "頁，沒有鞋子了")
            break

        for s in shoes:
            pic=s.find('img',class_='rt-product-image')['src']
            shoes_web=s.find('a')['href']
            shoes_name=s.find('h3',class_='item-name').text
            shoes_price=s.find('span',class_='item-direct-price').text.strip().replace("$", "")
            try:
                shoes_volume=s.find('div',class_='item-price').find_all('p')[1].text.strip()

            except IndexError:
                shoes_volume = None
    #         print(shoes_name, shoes_web, shoes_volume, pic)
            saved = {
                "picture": pic,
                "web": shoes_web,
                "name": shoes_name,
                "volum": shoes_volume,
                "price": shoes_price
            }
    #         print(json.dumps(saved, indent=4))

            if not os.path.exists(dn):
                os.makedirs(dn)

            fn= dn + pic.split('/')[-1].rsplit(".")[0] #檔案路徑+檔名:把檔案放在同目錄的ruten_miss_aprils資料夾裡
            # 儲存內文(JSON)
            f = open(fn + '.json','w',encoding='utf_8')
            json.dump(saved,f)
            f.close()
            urlretrieve(pic, dn + pic.split('/')[-1])
            time = random.randint(1,3)
            print("防阻擋時間:", time)
            sleep(time)
        page += 1
```
